```python
# Import dependencies/modules
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
# File to Load (Remember to Change These)
mouse_drug_data_to_load = "mouse_drug_data.csv"
clinical_trial_data_to_load = "clinical_trial_data.csv"
```

```python
# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data = pd.read_csv(clinical_trial_data_to_load)
```

```python
# Combine the data into a single dataset using Mouse_ID as common identifier
combined_trial_data = pd.merge(mouse_drug_data, clinical_trial_data,how='outer', on="Mouse ID")

# Display the data table for preview
combined_trial_data.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
combined_tumor_volume_mean = pd.DataFrame(combined_trial_data.groupby(['Drug', 'Timepoint']).mean()['Tumor Volume (mm3)'])
combined_tumor_volume_mean.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
combined_tumor_volume_stde = pd.DataFrame(combined_trial_data.groupby(['Drug', 'Timepoint']).sem()['Tumor Volume (mm3)'])
combined_tumor_volume_stde.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
table_tumor_volume_mean = combined_tumor_volume_mean.unstack(level = 0)
table_tumor_volume_mean.columns = table_tumor_volume_mean.columns.get_level_values(1)
table_tumor_volume_mean
```

```python
# Minor Data Munging to Re-Format the Data Frames
table_tumor_volume_stde = combined_tumor_volume_stde.unstack(level = 0)
table_tumor_volume_stde.columns = table_tumor_volume_stde.columns.get_level_values(1)
table_tumor_volume_stde
```

```python
combined_trial_data.describe()
```

```python
# Set x_axis
x_axis=np.arange(0,50,5)

#Set plot size
plt.figure(figsize = (15,10))

# Set error bars and line formatting
plt.errorbar(table_tumor_volume_mean.index, table_tumor_volume_mean["Capomulin"], yerr=table_tumor_volume_stde["Capomulin"], color="red", marker="d", markersize=5, linestyle="dashed", linewidth=0.75)
plt.errorbar(table_tumor_volume_mean.index, table_tumor_volume_mean["Infubinol"], yerr=table_tumor_volume_stde["Infubinol"], color="blue", marker="^", markersize=5, linestyle="dashed", linewidth=0.75)
plt.errorbar(table_tumor_volume_mean.index, table_tumor_volume_mean["Ketapril"], yerr=table_tumor_volume_stde["Ketapril"], color="green", marker="s", markersize=5, linestyle="dashed", linewidth=0.75)
plt.errorbar(table_tumor_volume_mean.index, table_tumor_volume_mean["Placebo"], yerr=table_tumor_volume_stde["Placebo"], color="black", marker="o", markersize=5, linestyle="dashed", linewidth=0.75)

# Set descriptors 
plt.title("Tumor Response to Treatment", fontsize=18)
plt.xlabel("Treatment Days", fontsize=12)
plt.ylabel("Tumor Volume (mm3)", fontsize=12)
plt.grid(linestyle="dotted")
plt.legend(("Capomulin", "Infubinol", "Ketapril", "Placebo"), fontsize=12)

# Save the Figure
plt.savefig("../TumorVolumeResponse.png")

# Show the Figure
plt.show()
```

```python
# Metastatic Response to Treatment
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
metastatic_response_mean = pd.DataFrame(combined_trial_data.groupby(['Drug', 'Timepoint']).mean()['Metastatic Sites'])
metastatic_response_mean.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
metastatic_response_stde = pd.DataFrame(combined_trial_data.groupby(['Drug', 'Timepoint']).sem()['Metastatic Sites'])
metastatic_response_stde.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
table_metastatic_response_mean = metastatic_response_mean.unstack(level = 0)
table_metastatic_response_mean.columns = table_metastatic_response_mean.columns.get_level_values(1)
table_metastatic_response_mean
```

```python
# Minor Data Munging to Re-Format the Data Frames
table_metastatic_response_stde = metastatic_response_stde.unstack(level = 0)
table_metastatic_response_stde.columns = table_metastatic_response_stde.columns.get_level_values(1)
table_metastatic_response_stde
```

```python
# Generate the Plot (with Error Bars)

# Set x_axis
x_axis=np.arange(0,50,5)

#Set plot size
plt.figure(figsize = (15,10))

# Set error bars and line formatting
plt.errorbar(table_metastatic_response_mean.index, table_metastatic_response_mean["Capomulin"], yerr=table_metastatic_response_stde["Capomulin"], color="red", marker="d", markersize=5, linestyle="dashed", linewidth=0.75)
plt.errorbar(table_metastatic_response_mean.index, table_metastatic_response_mean["Infubinol"], yerr=table_metastatic_response_stde["Infubinol"], color="blue", marker="^", markersize=5, linestyle="dashed", linewidth=0.75)
plt.errorbar(table_metastatic_response_mean.index, table_metastatic_response_mean["Ketapril"], yerr=table_metastatic_response_stde["Ketapril"], color="green", marker="s", markersize=5, linestyle="dashed", linewidth=0.75)
plt.errorbar(table_metastatic_response_mean.index, table_metastatic_response_mean["Placebo"], yerr=table_metastatic_response_stde["Placebo"], color="black", marker="o", markersize=5, linestyle="dashed", linewidth=0.75)

# Set descriptors 
plt.title("Number of Metastatic Sites", fontsize=18)
plt.xlabel("Treatment Days", fontsize=12)
plt.ylabel("Metastatic Sites", fontsize=12)
plt.grid(linestyle="dotted")
plt.legend(("Capomulin", "Infubinol", "Ketapril", "Placebo"), fontsize=12)

# Save the Figure
plt.savefig("../MetastaticResponse.png")

# Show the Figure
plt.show()
```

```python
# Survival Rates
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
number_mice = pd.DataFrame(combined_trial_data.groupby(['Drug', 'Timepoint']).count()['Mouse ID'])
number_mice.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
table_number_mice = number_mice.unstack(level = 0)
table_number_mice.columns = table_number_mice.columns.get_level_values(1)
table_number_mice
```

```python
# Generate the Plot (Accounting for percentages)

# Set x_axis
x_axis=np.arange(0,50,5)

#Set plot size
plt.figure(figsize = (15,10))

# Set error bars and line formatting
plt.errorbar(table_number_mice.index, (table_number_mice["Capomulin"]/25*100), color="red", marker="d", markersize=5, linestyle="dashed", linewidth=0.75)
plt.errorbar(table_number_mice.index, (table_number_mice["Infubinol"]/25*100), color="blue", marker="^", markersize=5, linestyle="dashed", linewidth=0.75)
plt.errorbar(table_number_mice.index, (table_number_mice["Ketapril"]/25*100), color="green", marker="s", markersize=5, linestyle="dashed", linewidth=0.75)
plt.errorbar(table_number_mice.index, (table_number_mice["Placebo"]/25*100), color="black", marker="o", markersize=5, linestyle="dashed", linewidth=0.75)

# Set descriptors 
plt.title("Survival Rate thru Treatment", fontsize=18)
plt.xlabel("Treatment Days", fontsize=12)
plt.ylabel("Survival Rate (%)", fontsize=12)
plt.grid(linestyle="dotted")
plt.legend(("Capomulin", "Infubinol", "Ketapril", "Placebo"), fontsize=12)

# Save the Figure
plt.savefig("../SurvivalRate.png")

# Show the Figure
plt.show()
```

```python
# Summary Bar Graph that compares the total % tumor volume change for each drug across the full 45 days.

# Calculate the percent changes for each drug
tumor_vol_change =  100 * (table_tumor_volume_mean.iloc[-1] - table_tumor_volume_mean.iloc[0]) / table_tumor_volume_mean.iloc[0]
tumor_vol_change_stde =  100 * (table_tumor_volume_stde.iloc[-1] - table_tumor_volume_stde.iloc[0]) / table_tumor_volume_stde.iloc[0]
tumor_vol_change
```

```python
# Store all Relevant Percent Changes into a Tuple
tumor_percent_changes = (tumor_vol_change["Capomulin"],
                         tumor_vol_change["Infubinol"],
                         tumor_vol_change["Ketapril"],
                         tumor_vol_change["Placebo"])
tumor_percent_changes
```

```python
df = pd.DataFrame({'Drug':['Capomulin', 'Infubinol', 'Ketapril', 'Placebo'], 'val':[-19.475303, 46.123472, 57.028795, 51.297960]})
ax = df.plot.bar(x='Drug', y='val', rot=0)
```

```python
# Splice the data between passing and failing drugs 
plt.figure(figsize = (15,10))

fig, ax = plt.subplots()
ind = np.arange(len(tumor_percent_changes))
width = 0.85
rectsReduction = ax.bar(ind[0], tumor_percent_changes[0], width, color='green')
rectsGrowth = ax.bar(ind[1:], tumor_percent_changes[1:], width, color='red')

# Orient widths. Add labels, tick marks, etc.  
ax.set_ylabel('Tumor Volume Change (%)', fontsize=12)
ax.set_title('Tumor Change Throughout Treatment', fontsize=18)
ax.set_xticks(ind)
ax.set_xticklabels(('Capomulin', 'Infubinol', 'Ketapril', 'Placebo'), fontsize=12)
ax.set_xlim(-0.5, 3.5)
ax.set_ylim([-30,70])
plt.grid(linestyle='dotted',alpha=0.5)
plt.axhline(y=0,color='k')

# Use functions to label the percentages of changes
def autolabelReduction(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., -8,
                '-%d%% ' % int(height),
                ha='center', va='bottom', color="white")
def autolabelGrowth(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 3,
                '%d%%' % int(height),
                ha='center', va='bottom', color="black")

# Call functions to implement the function calls
autolabelReduction(rectsReduction)
autolabelGrowth(rectsGrowth)

# Save the Figure
plt.savefig("../TumorGrowthReduction.png")

# Show the Figure
fig.show()
```
